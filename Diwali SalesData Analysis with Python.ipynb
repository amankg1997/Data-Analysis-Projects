## Exploratory Data Analysis(EDA)

```python
df.columns
```

### Gender Column

```python
sns.set(rc = {'figure.figsize':(2,4)})
sns.countplot(x='Gender', data = df)
```

```python
gender_count = sns.countplot(x='Gender', data = df)
sns.set(rc = {'figure.figsize':(2,4)})
for bar in gender_count.containers:
    gender_count.bar_label(bar)
```

```python
## to groupby the data on the basis of Gender, want to get the total amount, sorting this by Amount in descending order
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
```

```python
### to get the above data in graphical form
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc = {'figure.figsize':(3,4)})
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

### Age Column

```python
df.columns
```

```python
sns.set(rc = {'figure.figsize':(7,6)})
sns.countplot(x='Age Group', hue = 'Gender', data=df)
```

```python
salescount_age = sns.countplot(x='Age Group', hue = 'Gender', data=df)
for bars in salescount_age.containers:
    salescount_age.bar_label(bars)
```

```python
sales_by_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Age Group', y= 'Amount', data = sales_by_age)
```

### Conclusion: By above data, we observed that most of the people who are buying products is of age group between 26-35 years and women are buying more than men.

### State Column

```python
sales_by_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(20,8)})
sns.barplot(data = sales_by_state, x='State', y = 'Orders')
```

### From the above data we can see that most number of orders are done from UP, Maharashtra and Karnataka respectively.

```python
sales_by_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(21,9)})
sns.barplot(data = sales_by_state, x='State', y='Amount')
```

### Insights: Most of the orders and sales are from UP, Maharashtra and Karnataka

```python
married_count = sns.countplot(data=df, x='Marital_Status')
sns.set({'figure.figsize':(4,6)})
for bars in married_count.containers:
    married_count.bar_label(bars)
```

```python
sales_by_marriage = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(4,6)})
sns.barplot(data = sales_by_marriage, x='Marital_Status', y='Amount', hue='Gender')
```

### Insights: As per the above data we observed that women are buying more than men

### Occupation Column

```python
sns.set(rc={'figure.figsize':(10,5)})
plt.xticks(rotation=45)
occup_count = sns.countplot(x='Occupation', data=df)

for bars in occup_count.containers:
    occup_count.bar_label(bars)
```

```python
sales_by_occup = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc = {'figure.figsize':(10,4)})
plt.xticks(rotation=45)
sns.barplot(data = sales_by_occup, x='Occupation', y='Amount')
```

### From the above data we observed that most of the buyers are from IT sector, Healthcare, and Aviation.

### Product_Category

```python
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation=45)
categ_count = sns.countplot(data=df, x='Product_Category')
for bars in categ_count.containers:
    categ_count.bar_label(bars)
```

```python
sales_by_categ = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc = {'figure.figsize':(20,5)})
plt.xticks(rotation=45)
sns.barplot(data = sales_by_categ, x='Product_Category', y = 'Amount')
```

### From the above data, most buying products are from Food, Clothing and Electronics category.  

```python
sales_by_orders = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc = {'figure.figsize':(20,5)})
plt.xticks(rotation=45)
sns.barplot(data = sales_by_orders, x='Product_ID', y = 'Orders')
```

### From the above data, most selling product id are: P00265242, P00110942, and P00237542

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
